```python
#obtener valores de la curva, solo un frame de momento

import pandas as pd
import numpy as np
import os
import pandas as pd
import dlisio
import time
start_time = time.time()

curves = []
curve_min = []
curve_max = []
curve_mean = []
curve_75 = []
curve_25 = []
nombre_curva = []
nombre_largo = []
units = []
min_frame = []
max_frame = []
name_producer = []
pozo = []
campo = []
compania = []
fecha_elaboracion = []
soft_ver = []
version_soft = []
file_id = []
compania_servicio = []
nombre_frame = []
archivo = []
archivos = []
uni = []
lo = []
files_L = []
curves_nam = []
filesitos = []
we = []
direccion = []

folderpath = r'C:\Users\jrogu\Desktop\Azzait\Prueba_todos\APKMZ\APKMZ\ZAAP\ORIGINALES\ZAAP-51'

for (root, dirs, files) in os.walk(folderpath):
    for f in files:
        filepath = os.path.join(root, f)
        if filepath.endswith('.' + 'dlis'):
            with dlisio.load(filepath) as files:
                for d in files:
                    for fram in d.frames:
                        files_L.append(d)
                        curves = fram.curves()
                        for file in d.origins:
                            for curva,i in enumerate (curves.dtype.names[0:]):
                                for channel in d.channels:
                                    if curves.dtype.names[curva] == channel.name:
                                        nombre_curva.append(curves.dtype.names[curva])
                                        curve_min.append(np.nanmin(curves[i]))
                                        curve_max.append(np.nanmax(curves[i]))
                                        curve_mean.append(np.nanmean(curves[i]))
                                        curve_75.append(np.nanpercentile(curves[i],75))
                                        curve_25.append(np.nanpercentile(curves[i],25))
                                        min_frame.append(((fram.index_min/10) / 39.37))
                                        max_frame.append(((fram.index_max/10) / 39.37))
                                        name_producer.append(file.producer_name)
                                        direccion.append(fram.direction)
                                        pozo.append(file.well_name)
                                        campo.append(file.field_name)
                                        compania.append(file.company)
                                        compania_servicio.append(file.producer_name)
                                        fecha_elaboracion.append(file.creation_time)
                                        soft_ver.append(file.product)
                                        version_soft.append(file.version)
                                        file_id.append(file.file_id)
                                        nombre_frame.append(fram.name)
                                        nombre_largo.append(channel.long_name)
                                        units.append(channel.units)
                                        #archivo.append(filepath)
                                        uno = os.path.split(filepath)
                                        archivo.append(uno[1])

curvita = pd.DataFrame(
{'Curva': nombre_curva,
 'Nombre largo': nombre_largo,
 'Unidad': units,
 'Minimo': curve_min,
'Maximo': curve_max,
'Media': curve_mean,
'Percentil 75': curve_75,
'Percentil 25': curve_25,
 'Nombre Frame': nombre_frame,
 'Direccion': direccion,
 'Min Frame': min_frame,
 'Max Frame': max_frame,
 'Pozo': pozo,
 'Campo': campo,
 'Compania' : compania,
 'Fecha': fecha_elaboracion,
 'Version Software':soft_ver,
 'File Id': file_id,
 'Elaboro': name_producer,
 'Archivo': archivo,
})

curvita.to_excel(r'C:\dataset\demo\APKMZ_DLIS.xlsx')

print("--- %s seconds ---" % (time.time() - start_time))
```
